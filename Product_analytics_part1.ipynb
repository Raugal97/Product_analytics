# Коллега прислал вам следующие данные:

ga:date – дата
ga:clientid – идентификатор пользователя (из основной системы аналитики)
userID – идентификатор пользователя (из другой системы)
ga:transaction_id – id транзакции
ga:revenue – доход
ga:user – колонка с единственной ячейкой, в ней хранится число уникальных пользователей

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/lesson/423116/august_data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

# Сколько всего имеется уникальных пользователей?

```python
uniq_users = df['ga:user '].max()

uniq_users
```

```python
df = df.drop(df.columns[[5, 6]], axis=1)

df = df.rename(columns={'ga:date': 'date',
                        'ga:clientid': 'client_id',
                        'userID': 'user_id',
                        'ga:transaction_id': 'transaction_id',
                        'ga:revenue': 'revenue'})
df['client_id'] = df['client_id'].astype(np.int64)
df['date'] =  pd.to_datetime(df['date'], format='%d-%m-%Y %H:%M:%S')
df['user_id'] = df['user_id'].str.replace(',', '')
```

```python
df.head()
```

# Сколько всего имеется клиентов?

```python
number_of_clients = df.query('revenue != 0') \
                        .client_id \
                        .nunique()

number_of_clients
```

# Какая конверсия?

```python
conversion_percent = round((number_of_clients / uniq_users) * 100, 2)

conversion_percent
```

# Сколько рублей составляет средний чек?

```python
average_check = round(df['revenue'].sum() / df.query('revenue != 0')['transaction_id'].count())

average_check
```

# Сколько повторных закупок? (среднее количество покупок на пользователя)

```python
average_amount_by_client = round(df.query('revenue != 0')['transaction_id'].count() / number_of_clients, 2)

average_amount_by_client
```

# Чему равен средний доход на платящего пользователя (ARPPU)?

```python
ARPPU = round(df['revenue'].sum() / number_of_clients)

ARPPU
```

# Посчитайте средний доход с пользователя (ARPU)

```python
ARPU = round((number_of_clients / uniq_users) * (df['revenue'].sum() / number_of_clients))

ARPU
```

```python
ARPU_v2 = round(df['revenue'].sum() / uniq_users)

ARPU_v2
```
